### Exploratory data analysis (EDA) 
**- EDA is used by data scientists to analyze and investigate data sets and summarize their main characteristics, often employing data visualization methods.**

**Essential Steps in EDA:**

- Data Cleaning
- Data Summary and visualization
    - Descriptive Statistics: Calculate basic statistics like mean, median, mode, standard deviation, and variance to summarize and describe the data.
    - Data visualization: Create visual representations of the data, such as bar charts, histograms, scatter plots, and box plots, to help identify patterns and relationships.
        - Distributions
        - Outliers
    - Handling Missing Data
    - Data formatting
    - Transformation
- Encoding and featurization (feature engineering)

**Exploratory data analysis tools**:
    
- **Clustering and dimension reduction** techniques, which help create graphical displays of high-dimensional data containing many variables.
- **Univariate visualization** of each field in the raw dataset, with summary statistics.
- **Bivariate visualizations** and summary statistics that allow you to assess the relationship between each variable in the dataset and the target variable you’re looking at.
- **Multivariate visualizations**, for mapping and understanding interactions between different fields in the data.

**Grouping data values into bins**

`pd.cut`
**Bin values into discrete intervals.**

Use cut when you need to segment and sort data values into bins. This function is also useful for going from a continuous variable to a categorical variable. For example, cut could convert ages to groups of age ranges. Supports binning into an equal number of bins, or a pre-specified array of bins.

```python
my_bins = [0, 100, 500, 1000, np.inf]
my_labels=['Not Popular','Slightly Popular','Popular', 'Very Popular']
df['popularity_rank'] = pd.cut(df['popularity'], bins=my_bins, labels=my_labels)
```

```python
# More feature engineering for the Genre
gen_col = []
def get_all_genres(genres):
    # pick each genre and split
    for each in genres.split(','):
        each = each.strip()
        gen_col.append(each)
    return np.nan
```

```python
t = df['genre_names'].apply(get_all_genres)
# now you have all the genre that are in you dataset
gen_col = list(set(gen_col))
print(gen_col)
```

```python
#  a function that split each set of genre per row and convert to digital form
def label_gen(x, label):
    x = x.strip()
    # if the genere we are on can be found in set of genre in a row
    if label in x:
        return 1
    else:
        return 0
```

```python
for each_add in gen_col:
    print(each_add)
```

```python
for each_add in gen_col:
    df[each_add] = df['genre_names'].apply(lambda x:label_gen(x, each_add))
```

### Clustering

```python
#numeric data
my_numeric_df = df.select_dtypes(exclude ='object') # obtains the numerical data type
my_numeric_df
```

```python
ad_dummies = pd.get_dummies(my_numeric_df.adult)
my_numeric_df = pd.concat([my_numeric_df, ad_dummies], axis=1)
```

```python
my_numeric_df.drop(['release_date','adult'], axis=1, inplace=True)
```

```python
my_numeric_df
```

```python
my_numeric_df.columns
```

```python
!pip install KElbowVisualizer
```

```python
from sklearn.compose import make_column_transformer,ColumnTransformer, make_column_transformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder, LabelEncoder
```

```python
# Prepare function that will preprocess additional features
normalize = MinMaxScaler()
one_hot = OneHotEncoder()
label_code = LabelEncoder()
```

```python
ct = make_column_transformer(
    (one_hot , ['popularity_rank']),
    remainder = 'passthrough'
)
extracted_features = ct.fit_transform(my_numeric_df)
```

```python
extracted_features
```

```python
visualizer3 = KElbowVisualizer(KMeans(), k=(2,11))
visualizer3.fit(extracted_features) # Fit the data to the visualizer
visualizer3.show(outpath='elbowplot2.png')   
```

```python
#Create and train (fit) the model

model = KMeans(n_clusters=4, random_state=42)
model = model.fit(extracted_features)
```

```python
#Get cluster centres (centroids)
centroids = model.cluster_centers_
```

```python
model.labels_
```

```python
df['grp'] = model.labels_ + 1
```

```python
df
```

```python

sns.lmplot(data=df, x='year', y='vote_average', hue='grp', markers=["o", "x", "1", "*"], fit_reg=False)
```
