### Search Movie

```python
selected_movie = input("Type Movie Name : ")
# Orphan: First Kill
```

```python
def searcher(mvs_name, kw):
    recomd = []
    if kw in mvs_name.split():
        recomd.append(mvs_name)
    if len(recomd)>0:
        return recomd
```

```python
rcd = df['title'].apply(lambda x:searcher(x, selected_movie))
for each in rcd:
    if each != None:
        print(each)
```
